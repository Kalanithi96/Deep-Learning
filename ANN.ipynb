# Importing Libraries

```python
import numpy as np
import tensorflow as tf
import pandas as pd
import math
```

# Part 1 - Data Preprocessing

Starting from column [3] to One before Last [ 3 : -1 ]

```python
dataset = pd.read_csv('/content/Churn_Modelling.csv')
x = dataset.iloc[ : , 3:-1].values
y = dataset.iloc[ : , -1].values
```

```python
print(x)
```

```python
print(y)
```

Label encoding : Gender

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[ : , 2] = le.fit_transform(x[ : , 2])
```

```python
print(x)
```

Label encoding "Geography Column"

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer( transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```python
print(x)
```

Splitting the dataset into Training set and Testing set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y ,test_size = 0.2, random_state = 0)
```

Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

# Part 2 - Building the ANN

Initializing the ANN

Tf contains Keras contains Models contains Sequential

ann is an object of type Sequential

```python
ann = tf.keras.models.Sequential()
```

Adding the input layer and the first hidden layer

Dense() class is a hidden layer that takes all the nodes in previous layer as input

Parameters of Dense():

units -> number of nodes in the layer [Hyperparameter]

activation -> activation function

"relu" - rectification activation function"

"sigmoid" - sigmoid activation function" for Binary classification Output

"softmax" - for Multiclass classification Output

**Get Activation Functions**

**Get Layer Types**

```python
ann.add( tf.keras.layers.Dense( units=6, activation='relu' ) )
```

Adding the second hidden layer

```python
ann.add( tf.keras.layers.Dense( units=6, activation='relu' ) )
```

Adding the Output layer

The output is Binary outcome: so output units is 1

If outcome is Multiclass, eg n=3 We would need 3 output nodes

```python
ann.add( tf.keras.layers.Dense( units=1, activation='sigmoid' ) )
```

# Part 3 - Training the ANN

Compiling the ANN

Compile()

Parameters:

optimizer = 

"adam" -> Stochastic Gradient Descent

loss = Loss Function

'binary_crossentropy' -> Binary classification

'categorical_crossentropy' -> Multi-class Classification

metrics = []

'accuracy' - Accuracy score

**Get Optimizers**

**Get Loss functions**

**Get Metrics**

```python
ann.compile( optimizer="adam" , loss='binary_crossentropy'  , metrics=[ 'accuracy' ] )
```

Training the ANN on the Training set

batch_size -> number of rows in a Batch

epochs -> Iterations

```python
ann.fit(x_train, y_train, batch_size=32, epochs=100 )
```

# Part 4 - Making the Predictions and evaluating the model

Make a Prediction:

Geography = France

Gender = Male

Age = 40 yrs old

Tenure = 3 yrs

Balance = $60000

```python
print(ann.predict( sc.transform([[1, 0, 0, 600, 1, 40, 3, 60000, 2, 1, 1, 50000]] )) > 0.5 )
```

1. Input of ANN is a 2D array

2. Encode France before Input

    First Three values => Code for France

3. Apply the same scaling. 

    Using sc.transform()

4. Output will be a Probability.

    Using Boolean Operator



Predicting the Test results

```python
y_pred = ann.predict(x_test)
y_pred = ( y_pred > 0.5 )
print(np.concatenate( 
                        (   y_pred.reshape( len(y_pred),1 ), 
                            y_test.reshape( len(y_test),1 ) 
                        ),
                        1 
                     )
      )
```

Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

# Custom Activation Function

```python
import keras.backend as K
from keras.layers import Activation
from keras.utils.generic_utils import get_custom_objects

def custom_activation(x):
  return K.maximum(K.zeros_like(x),tf.math.log(1+x))
```

```python
custom8 = tf.keras.models.Sequential()
custom8.add( tf.keras.layers.Dense( units=6, activation=custom_activation ) )
custom8.add( tf.keras.layers.Dense( units=1, activation='sigmoid' ) )
```

```python
custom8.compile( optimizer="adam" , loss='binary_crossentropy'  , metrics=[ 'accuracy' ] )
custom8.fit(x_train, y_train, batch_size=32, epochs=100 )
```

```python
y_pred = custom8.predict(x_test)
y_pred = ( y_pred > 0.5 )
print(np.concatenate( 
                        (   y_pred.reshape( len(y_pred),1 ), 
                            y_test.reshape( len(y_test),1 ) 
                        ),
                        1 
                     )
      )

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
